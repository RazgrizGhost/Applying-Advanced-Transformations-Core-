```python
# Imports
import numpy as np
import pandas as pd
import ast

# Import OS and JSON 
import os,json
```

```python
superdf = pd.read_csv("Data/superhero_info - superhero_info.csv")
display(superdf)
superdf.info()
```

```python
superpowerdf = pd.read_csv("Data/superhero_powers - superhero_powers.csv")
display(superpowerdf)
superpowerdf.info()
```

```python
superdf["Hero|Publisher"].value_counts()
```

```python
superdf["Hero|Publisher"].str.split('|', expand = True)
```

```python
# Seperated the Hero|Publisher Column into two different columns

superdf[["Hero", "Publisher"]] = superdf["Hero|Publisher"].str.split('|', expand = True)
superdf = superdf.drop(columns = "Hero|Publisher")
superdf
```

```python
# replacing all single quotes for json.loads
superdf["Measurements"] = superdf["Measurements"].str.replace("'",'"')

# applying json.loads for the whole column
superdf["Measurements"] = superdf["Measurements"].apply(json.loads)
superdf["Measurements"]
```

```python
# making the columns for Height and Weight
h_w = superdf["Measurements"].apply(pd.Series)
h_w
```

```python
# combining it with the main DataFrame
superdf = pd.concat((superdf, h_w), axis = 1)
superdf = superdf.drop(columns = "Measurements")
superdf
```

```python
superpowerdf["Powers_Split"] = superpowerdf["Powers"].str.split(",") 
superpowerdf["Powers_Split"]
```

```python
exploded = superpowerdf.explode("Powers_Split")
exploded[["hero_names", "Powers_Split"]]
```

```python
cols_to_make = exploded["Powers_Split"].dropna().unique()
cols_to_make
```

```python
# For Loop for each of every Powers and making them into columns
for col in cols_to_make:
    superpowerdf[col] = superpowerdf["Powers_Split"].str.contains(col, regex = False)
superpowerdf.head()
```

```python
superpowerdf = superpowerdf.drop(columns = ["Powers", "Powers_Split"])
superpowerdf
```

```python
# Merging DataFrames
merged_df = pd.merge(superdf, superpowerdf, left_on = "Hero", right_on = "hero_names")
merged_df
```

# Data Cleaning

```python
merged_df["Hair color"].value_counts()
```

```python
merged_df["Hair color"]= merged_df["Hair color"].replace("blond", "Blond").replace("black", "Black").replace("Brownn", 
                                                                                                             "Brown")
merged_df["Hair color"].value_counts()
```

```python
merged_df["Height"] = merged_df["Height"].replace(" cm", "", regex = True)
merged_df["Height"]
```

```python
merged_df["Weight"] = merged_df["Weight"].replace(" kg", "", regex = True)
merged_df["Weight"]
```

```python
merged_df["Weight"] = merged_df["Weight"].astype(float)
merged_df["Height"] = merged_df["Height"].astype(float)
```

```python
merged_df.to_csv("SuperPowersDataFrame", index = False)
```

1. Compare the average weight of super powers who have Super Speed to those who do not. 

```python
SuperSpeedFilter = merged_df["Super Speed"] == True
SuperSpeed = merged_df[SuperSpeedFilter]
SuperSpeedWeight = SuperSpeed["Weight"].mean()
```

```python
NonSuperSpeedFilter = merged_df["Super Speed"] == False
NonSuperSpeed = merged_df[NonSuperSpeedFilter]
NonSuperSpeedWeight = NonSuperSpeed["Weight"].mean()
```

```python
print(f"The Average Super Speed Heroes and Non Super Speed Weight. SuperSpeed: {SuperSpeedWeight} NonSuperSpeed: {NonSuperSpeedWeight}")
```

2. What is the average height of heroes for each publisher?

```python
merged_df.groupby(["Publisher"])["Height"].mean()
```
